```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pandas as pd
```

```python
df = pd.read_csv('fundamentals.csv', usecols = ['Fixed Assets', 'Gross Profit'])
df.head()
```

```python
x = df[['Fixed Assets']]
y = df['Gross Profit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_predictions = regressor.predict(X_test)
score = r2_score(y_test, y_predictions)
print('SSE:', score)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

plt.scatter(X_train, y_train, color='black')
plt.plot(X_train, regressor.predict(X_train),color='y')

plt.show()
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial = PolynomialFeatures(degree=4)
X_polynomial = polynomial.fit_transform(X_train)
regressor = LinearRegression()
regressor.fit(X_polynomial, y_train)
```

```python
plt.scatter(X_train, y_train, color='black')
plt.plot(X_train, regressor.predict(X_polynomial),color='y')
plt.show()
```
